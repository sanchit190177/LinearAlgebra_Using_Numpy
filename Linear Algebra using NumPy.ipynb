# **. System of Linear EQNS**
##### ***In this code, a system of linear equations is being solved using matrix operations in Python with the NumPy library.***

##### ***Breakdown of the Steps***
##### ***Defining Matrices:***

##### ***A 3x3 matrix named A is created from the numbers 1 to 8 and then 10. This matrix represents the coefficients of the variables.***

##### ***A 3x1 matrix (a column vector) named B is created. This matrix represents the constants on the right-hand side of the*** equations.

##### ***Solving the System:***

##### ***The code calculates the inverse of matrix A using np.linalg.inv(A).***

##### ***It then performs a matrix multiplication (@) of the inverse of A with matrix B.***

##### ***This operation, based on the formula x=A âˆ’1b, directly solves for the unknown variables. The result is stored in the variable X***



```python
A=np.array([1,2,3,4,5,6,7,8,10]).reshape(3,3)
B=np.array([10,11,12]).reshape(3,1)
```

```python
X=np.linalg.inv(A)@ B
X
```

```python
x=np.linalg.solve(A,B)
x
```
